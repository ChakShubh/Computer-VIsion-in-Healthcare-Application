### Importing Necessary Modules and Functions

```python
#import open CV
import cv2

#import numpy
import numpy as np

#import pandas
import pandas as pd

#import matplotlib for plotting the images
import matplotlib.pyplot as plt

#import scikit-image
from skimage.util import random_noise   
from skimage.filters import median
from skimage import io,color, img_as_float, img_as_int

#import weiner filter
from scipy.signal.signaltools import wiener
```

### Loading Image into the Python Notebook

```python
# Load the input image
image = cv2.imread('C:\\Users\\Subhendu Chakrabarti\\Desktop\\normal-fundus.png')
plt.imshow(image)
plt.title("Original Image in BGR configuration")
```

We note that the image is obtained in BGR format. Hence, we need to convert it into RGB format before carrying out further processing.

### Conversion of BGR Image to RGB Image

```python
#convert to RGB image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image in RGB configuration")
```

We could successfully convert the BGR format to RGB Configuration. 

We now obtain the Gray Scale view of the Original Image. This view would enable us to execute the exercise in a better way.

### Generating Gray Scale Image

```python
#convert to Gray Scale Image
gray_image = io.imread('C:\\Users\\Subhendu Chakrabarti\\Desktop\\normal-fundus.png', as_gray=True)
io.imshow(gray_image)
plt.title("Original Image in Gray Scale View")
```

We could successfully obtain the Gray Scale Image from the Original Color Image. 

### Analyzing the Image

```python
print("Shape of the Image: ", image.shape)
```

We see that the image is (337, 448, 3) in shape

We now add Noises to the Images

## Addition of Noises to Images

### Guassian Noise

```python
# Addition of Guassian Noise to the Image
gaussian_image = random_noise(gray_image, mode="gaussian", seed=None, clip=True)

#plotting Original Image
plt.subplot(1,2,1)
io.imshow(gray_image)
plt.title("Original Image")

#plotting Image after addition of Guassian Noise to it 
plt.subplot(1,2,2)
io.imshow(gaussian_image)
plt.title("Gaussian Noise")
```

### Salt and Pepper Noise 

```python
# Addition of Salt and Pepper Noise to the Image
sandp = random_noise(gray_image, mode='s&p',amount=0.3)

#plotting Original Image
plt.subplot(1,2,1)
io.imshow(gray_image)
plt.title("Original Image")

#plotting Image after addition of Salt and Pepper Noise to it 
plt.subplot(1,2,2)
io.imshow(sandp)
plt.title("Salt and Pepper Noise")
```

### Poisson Noise

```python
# Addition of Poisson Noise to the Image
poisson_image = random_noise(gray_image, mode="poisson", seed=None, clip=True)

#plotting Original Image
plt.subplot(1,2,1)
io.imshow(gray_image)
plt.title("Original Image")

#plotting Image after addition of Poisson Noise to it 
plt.subplot(1,2,2)
io.imshow(poisson_image)
plt.title("Poisson Noise")
```

### Speckle Noise

```python
# Addition of Speckle Noise to the Image
speckle_image = random_noise(gray_image, mode="speckle", seed=None, clip=True)

#plotting Original Image
plt.subplot(1,2,1)
io.imshow(gray_image)
plt.title("Original Image")

#plotting Image after addition of Speckle Noise to it 
plt.subplot(1,2,2)
io.imshow(speckle_image)
plt.title("Speckle Noise")
```

### Inference

We could successfully add Noises (namely, Guassian Noise, Salt and Pepper Noise, Poisson Noise, Speckle Noise) to the image and display it.

## Application of Filters on Noisy Images

### Gaussian Filter

#### Gaussian Noise 

```python
# Application on Gaussian Noise
result = cv2.GaussianBlur(gaussian_image, (3,3), sigmaX=35, sigmaY=40)
io.imshow(result)
plt.title("Gaussian Filter on Gaussian Noise")
```

#### Salt and Pepper Noise

```python
# Application on Salt and Pepper Noise
result = cv2.GaussianBlur(sandp, (3,3), sigmaX=35, sigmaY=40)
io.imshow(result)
plt.title("Gaussian Filter on Salt and Pepper Noise")
```

#### Poisson Noise

```python
# Application on Poisson Noise
result = cv2.GaussianBlur(poisson_image, (3,3), sigmaX=35, sigmaY=40)
io.imshow(result)
plt.title("Gaussian Filter on Poisson Noise")
```

#### Speckle Noise

```python
# Application on Speckle Noise
result = cv2.GaussianBlur(speckle_image, (3,3), sigmaX=35, sigmaY=40)
io.imshow(result)
plt.title("Gaussian filter on Speckle Noise")
```

### Inference

We could successfully apply Guassian Filter to all the Noisy Images.

Here we could successfully denoise the images as a result of Guassian filter application on the Noisy Images. But it is noted that even after applying Guassian filters on the Noisy Images, we observe the presence of white spots on Guassian Noised, Salt and Pepper Noised and Speckle Noised Images. These white spots indicates the existence of Noise even after the application Filters on the Noisy Images. 

We can conclude that the Guassian Filter could successfully on a relative basis denoise the Poisson Noised Image in a better way when compared to the denoising of other Noisy Images. 

### Median Filter

```python
# Speckle Noise
img = np.float32(speckle_image)
result = cv2.medianBlur(img, 3)
io.imshow(result)
plt.title("Median Filter on Speckle noise")
```

### Inference

## Simple Thresholding Technique

In this Thresholding Technique, for every pixel present in the image, the same threshold value is applied.

If the pixel value is smaller than the threshold, it is assigned to Min Value (0), Else, it is set to a Max Value (1).

Syntax: **cv2.threshold(image, thresholdValue, maxVal, thresholdingTechnique)** 

    1. image: Input Image 
    2. thresholdValue: Threshold Value
    3. maxVal: Maximum value that can be assigned to a pixel 
    4. thresholdingTechnique: The type of thresholding to be applied

The various Simple Thresholding Techniques are:
    
    1. Binary Thresholding
    2. Binary Thresholding Inverted
    3. Truncated Thresholding
    4. Set to Zero Thresholding
    5. Set to Zero Thresholding Inverted

We now apply the different Simple Thresholding Techniques on the Image

### Binary Thresholding

```python
#Binary Thresholding
ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1)
plt.title("Binary Threshold")
```

### Binary Threshold Inverted

## Adaptive Thresholding Technique

The threshold value in this technique is calculated for smaller regions. This results in different threshold values for different regions (mostly due to variable ambient lighting) of the image.

This eliminates the problem of applying a uniform threshold value to an image having different ambient lighting at different parts of the image.

**Syntax: cv2.adaptiveThreshold(image, maxVal, adaptiveMethod, thresholdType, blocksize, constant)** 

    1. image: Input Image 
    2. maxVal: Maximum value that can be assigned to a pixel 
    3. Adaptive Method: Determines the Technique
    4. Threshold Type: Determine type of Threshold
    5. Block Size: Size of Pixel

The various Adaptive Thresholding Techniques are:
    
    1. Adaptive Mean Thresholding - mean of the neighborhood area minus the constant C
    2. Adaptive Gaussian Thresholding - gaussian-weighted sum of the neighbourhood values minus the constant C

We now apply the different Adaptive Thresholding Techniques on the Image

### Adaptive Mean Threshold

```python
#Adaptive Mean Thresholding
thresh6 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 199, 5)
plt.imshow(thresh6)
plt.title("Adaptive Mean")
```

### Adaptive Gaussian Threshold

```python
#Adaptive Gaussian Thresholding
thresh7 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 199, 5)
plt.imshow(thresh7)
plt.title("Adaptive Gaussian")
```

## Otsu Thresholding Technique

The threshold value is not calculated but determined here. For the Histogram representation of a Bimodal Image, the Threshold Value would lie somewhere in between the two peaks of the Histogram.

**Syntax: cv2.threshold(image, thresholdValue, maxVal, thresholdingTechnique)** 

    1. image: Input Image 
    2. thresholdValue: Threshold Value 
    3. maxVal: Maximum value that can be assigned to a pixel 
    4. Adaptive Method: Determines the Technique
    5. thresholdingTechnique: Type of Thresholding Applied

```python
#Otsu Thresholding
ret, thresh8 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(thresh8)
plt.title("Otsu Threshold")
```

### Varying Parameters and analyzing the Output

We now change the Traditional Thresholding Method from THRESH_BINARY to THRESH_TOZERO and observe the change in the output.

```python
#Otsu Thresholding
ret, thresh8 = cv2.threshold(image, 200, 255, cv2.THRESH_TOZERO + cv2.THRESH_OTSU)
plt.imshow(thresh8)
plt.title("Otsu Threshold")
```

### Inference

Upon changing the tranditional thresholding technique, we observe a clearer image (for Set to Zero Thresholding)
